## Realizando una simulacion de inteligencia artificial utilizando POO, tipo menu
# Problema 15: Soporte Tecnico

```python
# Programa de Gestión de Soporte Técnico con Simulación de IA
# Utilizando Programación Orientada a Objetos (POO)

class SoporteTecnico:
    def __init__(self):
        # El constructor inicializa nuestra "base de datos" interna (una lista de diccionarios) [5, 6]
        self.registros = []

    def agregar_soporte(self):
        print("\n--- Registrar Nuevo Soporte ---")
        # Entrada de datos interactiva para los atributos requeridos [7, 8]
        ticket = input("Sistema de tickets (ID): ")
        nivel = input("Nivel de soporte (1, 2 o 3): ")
        empresa = input("Nombre de la empresa: ")
        tiempo = input("Tiempo de respuesta (ej. 15 min): ")
        software = input("Software de diagnóstico: ")
        dispositivos = input("Dispositivos gestionados: ")

        # Almacenamos los atributos en un diccionario y lo agregamos a la lista [9]
        nuevo_registro = {
            "ticket": ticket,
            "nivel": nivel,
            "empresa": empresa,
            "tiempo": tiempo,
            "software": software,
            "dispositivos": dispositivos
        }
        self.registros.append(nuevo_registro)
        print(f"✅ Soporte para '{empresa}' agregado correctamente.")

    def listar_soporte(self):
        # Presentación de la información en forma de tabla bien alineada usando f-strings [10, 11]
        if not self.registros:
            print("\n⚠️ No hay registros disponibles.")
        else:
            print("\n" + "="*115)
            print(f"{'TICKET':<10} | {'NIVEL':<7} | {'EMPRESA':<20} | {'TIEMPO':<15} | {'SOFTWARE':<25} | {'DISPOSITIVOS':<12}")
            print("-" * 115)
            for r in self.registros:
                print(f"{r['ticket']:<10} | {r['nivel']:<7} | {r['empresa']:<20} | {r['tiempo']:<15} | {r['software']:<25} | {r['dispositivos']:<12}")
            print("="*115)

    def actualizar_soporte(self):
        # Busca un ticket y permite modificar sus valores [12]
        buscar = input("\nIngrese el ID del Ticket a actualizar: ")
        encontrado = False
        for r in self.registros:
            if r['ticket'] == buscar:
                print(f"Editando empresa: {r['empresa']}")
                r['nivel'] = input("Nuevo nivel de soporte: ")
                r['tiempo'] = input("Nuevo tiempo de respuesta: ")
                print("✅ Registro actualizado.")
                encontrado = True
                break
        if not encontrado:
            print("❌ Ticket no encontrado.")

    def borrar_soporte(self):
        # Elimina un registro de la lista por su ID de ticket [12]
        buscar = input("\nIngrese el ID del Ticket a borrar: ")
        for i in range(len(self.registros)):
            if self.registros[i]['ticket'] == buscar:
                del self.registros[i]
                print("🗑️ Registro eliminado exitosamente.")
                return
        print("❌ No se encontró el ticket para eliminar.")

    def simular_ia(self):
        # Lógica de IA para dar recomendaciones basadas en el nivel de soporte [8, 13]
        if not self.registros:
            print("\n🤖 IA: No hay datos para analizar.")
            return

        print("\n🤖 --- ANÁLISIS DE INTELIGENCIA ARTIFICIAL ---")
        for r in self.registros:
            print(f"Análisis para {r['empresa']}:")
            nivel = r['nivel']
            if nivel == "1":
                print("   > IA sugiere: Problema menor. Enviar manual de usuario automático.")
            elif nivel == "2":
                print("   > IA sugiere: Problema técnico intermedio. Asignar técnico remoto.")
            elif nivel == "3":
                print("   > IA sugiere: ¡ALERTA CRÍTICA! Enviar ingeniero a sitio de inmediato.")
            else:
                print("   > IA sugiere: Nivel desconocido. Verificar datos.")

# --- Lógica del Menú Interactivo ---
# Creamos una instancia única de la clase [14, 15]
mi_sistema = SoporteTecnico()

while True:
    print("\n" + "*"*40)
    print("   MENU DE SOPORTE TÉCNICO IA")
    print("*"*40)
    print("1. Agregar Soporte")
    print("2. Listar Soportes (Tabla)")
    print("3. Actualizar Soporte")
    print("4. Borrar Soporte")
    print("5. Analizador de IA")
    print("6. Salir")

    opcion = input("\nSeleccione una opción (1-6): ")

    if opcion == "1":
        mi_sistema.agregar_soporte()
    elif opcion == "2":
        mi_sistema.listar_soporte()
    elif opcion == "3":
        mi_sistema.actualizar_soporte()
    elif opcion == "4":
        mi_sistema.borrar_soporte()
    elif opcion == "5":
        mi_sistema.simular_ia()
    elif opcion == "6":
        print("🤖 IA: Cerrando sistema. ¡Hasta luego!")
        break
    else:
        print("❌ Opción inválida, intente de nuevo.")
```
